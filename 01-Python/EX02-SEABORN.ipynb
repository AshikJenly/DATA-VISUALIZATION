```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./DataSets/salaries.csv")
```

# 1) Box Plot

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. It displays the median, quartiles, and potential outliers, providing a concise summary of the data's central tendency and spread. Box plots are used to visualize and compare the distributions of different datasets or identify outliers in data.

```python
sns.boxplot(x=df["salary"])
```

## With More Than One Field

```python
sns.boxplot(x = 'rank', y = 'salary', data=df)
```

```python
sns.boxplot(x = 'rank', y = 'salary', hue='sex', data=df)
```

# 2) Swarm Plot

A swarm plot is a type of categorical scatter plot that displays individual data points along a categorical axis. It prevents overlap by spreading points in a visually pleasing manner. Swarm plots are used to visualize the distribution of data points within categories, making them suitable for exploring relationships and patterns in small to moderate-sized datasets.

```python
sns.swarmplot(x = 'rank', y = 'salary', data=df)
```

# 3) Strip Plot
A strip plot is a type of one-dimensional scatter plot that displays individual data points as dots along a single axis. It's often used for visualizing the distribution of data points within a category or to compare data across multiple categories. Strip plots are particularly useful for small to medium-sized datasets to show the density and distribution of data within categories.

```python
sns.stripplot(x = 'rank', y = 'salary', data=df)
```

# 4) Joint Plot 

A jointplot is a type of data visualization in seaborn, a Python data visualization library. It combines multiple plots to display the relationships between two variables, including a scatter plot, histograms, and a correlation coefficient. Jointplots are used to explore the bivariate relationship and correlations between two variables in a single, informative plot.

```python
sns.jointplot(x = 'salary', y = 'service', data=df)
```

### with distribution

```python
sns.jointplot(x = 'salary', y = 'service', data=df,kind='reg')
```

### with kde

```python
sns.jointplot(x = 'salary', y = 'service', data=df).plot_joint(sns.kdeplot,n_levels=6)
```

# 5) Heat Map

A heat map is a graphical representation of data where values are depicted as colors in a grid. It is used to visualize relationships, patterns, or variations in data. Heat maps are commonly employed in fields like data analysis, statistics, and biology to display complex data sets, such as correlation matrices, temperature variations, or concentration gradients, in an intuitive and visually informative way

```python
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt="d")

```
