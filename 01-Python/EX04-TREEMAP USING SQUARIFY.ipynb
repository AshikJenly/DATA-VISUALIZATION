```python
import squarify
import matplotlib.pyplot as plt
```

```python
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 15, 29, 12, 41]
colors= ["red","black","green","violet","yellow","blue"]
```

# Simple Tree

```python
squarify.plot(data,color=colors)
plt.axis('off')
```

# With Axis

```python
squarify.plot(data,color=colors,label=cars,text_kwargs=)
```
